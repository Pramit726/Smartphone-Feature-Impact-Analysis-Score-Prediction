```python
def extended_extractor(row):

  if row['extended_memory_available'] == 0:
    return np.nan
  else:
    if row['extended_memory'] == '1 TB':
      return 1024
    elif row['extended_memory'] == '512 GB':
      return 512
    elif row['extended_memory'] == '256 GB':
      return 256
    elif row['extended_memory'] == 'Not Specified':
      return np.nan
    elif row['extended_memory'] == 'Memory Card (Hybrid)':
      return np.nan
    elif row['extended_memory'] == '128 GB':
      return 128
    elif row['extended_memory'] == '2 TB':
      return 2048
    elif row['extended_memory'] == '32 GB':
      return 32
    elif row['extended_memory'] == '64 GB':
      return 64
    elif row['extended_memory'] == '1000 GB':
      return 1000
```

```python
def extended_extractor(row):

  if row['extended_memory_available'] == 0:
    return np.nan
  else:
    if row['extended_memory'] == 'Not Specified':
      return np.nan
    elif row['extended_memory'] == 'Memory Card (Hybrid)':
      return np.nan
    else:
      return row['extended_memory']
```

```python
def transform(text):

  if text == '1':
    return '1024'
  elif text == '2':
    return '2048'
  elif text == '1000':
    return '1024'
  else:
    return text
```

```python
def os_transform(text):

  if 'Memory' in text:
    return np.nan
  elif 'android' in text:
    return text
  elif 'ios' in text:
    return text
  else:
    return 'other'
```

```python
def clean_data(df):
    
    return (
        df.assign(
            processor_brand=lambda df_: (
                df_.processor_brand
                .str.replace('sanpdragon', 'snapdragon')
                .str.replace('apple', 'bionic')
                .str.replace('samsung', 'exynos')
                .str.replace('qualcomm', 'snapdragon')
                .str.replace('a13', 'bionic')
            ),
            fast_charging_available=lambda df_: (
                df_.fast_charging.pipe(lambda ser: np.where(ser.eq(-1), 0, 1))
            ),
            fast_charging=lambda df_: (
                df_['fast_charging'].apply(lambda x: np.nan if x in [0, -1] else x)
            )
        ).pipe(lambda df_: df_.assign(
            ram_capacity=np.where(df_['internal_memory'].isnull(), 4, df_['ram_capacity']),
            internal_memory=np.where(df_['internal_memory'].isnull(), 64, df_['internal_memory'])
        ))
        .assign(
            extended_memory_available=lambda df_: (
                df_['extended_memory'].pipe(lambda ser: np.where(ser.eq('0'), 0, 1))
            ),
            extended_upto = lambda df_: (
                df_.apply(extended_extractor,axis=1).str.replace('\u2009',' ').str.split(' ').str.get(0).apply(transform)
            ),
            os=lambda df_: (
                df_.os.apply(os_transform)
            ),
            primary_camera_front=lambda df_: (
                    np.where(df_['primary_camera_front'].eq('Main'), np.nan, df_['primary_camera_front']).astype(float)
            ),
            num_cores=lambda df_: (
                df_.num_cores
                .str.replace('Octa Core','8')
                .str.replace('Hexa Core','6')
                .str.replace('Quad Core','4')
                .astype(float)
            )
        )

        .drop(columns=['processor_name','extended_memory'])
    )
```

```python
df_clean_2 = clean_data(df)
```
